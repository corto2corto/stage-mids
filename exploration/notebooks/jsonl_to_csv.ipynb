```python
import os
import json
import pandas as pd
```

```python
def jsonl_to_csv(source, dest):
    with open(source, "r", encoding="utf-8") as f:
        entrees = [json.loads(ligne) for ligne in f]
    entrees = [e for e in entrees if e["type"] == "ok"]

    df = pd.json_normalize(entrees, record_path="urls", meta="sitemap")
    df.columns = ["url", "sitemap"]
    df[["sitemap", "url"]].to_csv(dest, encoding="utf-8", index=False)
```

```python
SOURCE = "../data/urls"
DEST = f"{SOURCE}/urls"
os.makedirs(DEST, exist_ok=True)

for fichier in (f for f in  os.listdir(SOURCE) if f.endswith(".jsonl")):
    chemin_source = f"{SOURCE}/{fichier}"
    chemin_dest = f"{DEST}/{fichier.replace('.jsonl', '.csv')}"

    jsonl_to_csv(chemin_source, chemin_dest)
```
